```python
import requests
import json
import time
```

```python
dictionary = {}
dictionary['objects'] = []
```

```python
offset = 0
while True:
    data = requests.get('https://fcd.terra.dev/v1/txs?offset={}&limit=100&account=terra1f89xq3qhu98v4jch4y5xcrkhl9gytrne99x74t'.format(offset))
    if data.status_code != 200:
        time.sleep(60)
        data = requests.get('https://fcd.terra.dev/v1/txs?offset={}&limit=100&account=terra1f89xq3qhu98v4jch4y5xcrkhl9gytrne99x74t'.format(offset))
    
    data_json = data.json()
    listOfItems = data_json['txs']
    
    for item in listOfItems:
        dictionary['objects'].append(item)
    
    if "next" in data_json:
        offset = data_json['next']
    else:
        break
```
